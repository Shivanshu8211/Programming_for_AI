### Importing important libraries and function

```python
from skimage.io import imread, imshow
from numpy import fliplr, flipud
from matplotlib import pyplot as plt
```

### 1 - Loading the image information into a numpy array

```python
img_read = imread('scenery.png',as_gray = False)
```

### 2 - Printing the number of pixels in the image

```python
# As size function gives the total number of elements in 3-d array so in order to get no. of pixels we have to divide it by 3
print("Total number of pixels in image are:",img_read.size//3)
```

### 3 Function for returning maximum value and maximum index

```python
def max(a,b,c):
    max_val = 0
    max_index = 0
    for i in range(a):
        for j in range(b):
            d = img_read[i][j][c]
            if d > max_val:
                max_val = d
                max_index = [i,j,c]
    return max_val,max_index
```

### 3.a Representing the given image as red and printing maximum intensity value and its index

```python
img_R = img_read.copy()
img_R[:,:,1] = 0
img_R[:,:,2] = 0
plt.imshow(img_R)
plt.show()
print("RED : ",max(873,1600,0))
```

### 3.b Representing the given image as Green and printing maximum intensity value and its index

```python
img_G = img_read.copy()
img_G[:,:,0] = 0
img_G[:,:,2] = 0
plt.imshow(img_G)
plt.show()
print("GREEN : ",max(873,1600,1))
```

### 3.c Representing the given image as Blue and printing maximum intensity value and its index

```python
img_B = img_read.copy()
img_B[:,:,0] = 0
img_B[:,:,1] = 0
plt.imshow(img_B)
plt.show()
print("BLUE : ",max(873,1600,2))
```

### 4.a - Converting image into Gray scale

```python
new_img = imread('scenery.png',as_gray = True)
plt.subplot(121), imshow(img_read)
plt.subplot(122), imshow(new_img)
plt.show()
```

### 4.b - Flipping image horizontally

```python
H_flip_img = flipud(img_read)
plt.subplot(121), imshow(img_read)
plt.subplot(122), imshow(H_flip_img)
plt.show()
```

### 4.c - Flipping iamge vertiocally

```python
L_flip_img = fliplr(img_read)
plt.subplot(121), imshow(img_read)
plt.subplot(122), imshow(L_flip_img)
plt.show()
```
